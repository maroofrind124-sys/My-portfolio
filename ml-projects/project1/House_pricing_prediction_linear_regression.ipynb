# **Feature Engineering and Extraction**

```python
df_cleaned.head()
```

```python
df_cleaned['age'].value_counts()
```

```python
df_cleaned['lstat_log'] = np.log1p(df_cleaned['lstat'])
```

```python
df_cleaned['rm_squared'] = df_cleaned['rm'] ** 2
```

```python
df_cleaned['crim_log'] = np.log1p(df_cleaned['crim'])
df_cleaned['dis_log'] = np.log1p(df_cleaned['dis'])
```

```python
df_cleaned['tax_per_rad'] = df_cleaned['tax'] / df_cleaned['rad']
```

```python
df_cleaned['age_category'] = pd.cut(df_cleaned['age'],
                                     bins=[0,20,40,60,80,100],
                                     labels=['new','recent','mid','old','very_old'])
```

```python
df_cleaned.head()
```

```python
df_cleaned = pd.get_dummies(
    df_cleaned,
    columns=['age_category',]
)
```

```python
df_cleaned.head()
```

```python
bool_cols = ["age_category_new", "age_category_recent","age_category_mid","age_category_old","age_category_very_old"]
df_cleaned[bool_cols] = df_cleaned[bool_cols].astype(int)
```

# **Building Model**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
x=df_final.drop('medv',axis=1)
y=df_final['medv']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

**Standard Scaling**

```python
col=['zn','indus','nox','rad','tax','ptratio','lstat_log','rm_squared','crim_log','dis_log']
x_train=x_train.copy()
x_test=x_test.copy()

scaler=StandardScaler()
x_train[col] = scaler.fit_transform(x_train[col])
x_test[col] = scaler.transform(x_test[col])

model = LinearRegression()
model.fit(x_train, y_train)
y_predict = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
r2=r2_score(y_test,y_predict)
print(f"R2 value is : {r2}")
```

```python
n=x_test.shape[0]
p=x_test.shape[1]
adjusted_r2=1-((1-r2)*(n-1)/(n-p-1))
print(f"Adjusted r2 : {adjusted_r2}")
```

```python
print(f"Mean Absolute Error (MAE): {mean_absolute_error(y_test, y_predict):.2f}")
print(f"Root Mean Squared Error (RMSE): {np.sqrt(mean_squared_error(y_test, y_predict)):.2f}")
```

**Results Visualization**

```python
plt.figure(figsize=(5,5))
plt.scatter(y_test, y_predict, alpha=0.6, color='#4A90A4', edgecolor='k', s=50)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()],
         'r--', lw=2, label='Perfect Prediction (y = x)')
plt.xlabel('Actual Price ($1000s)', fontsize=12)
plt.ylabel('Predicted Price ($1000s)', fontsize=12)
plt.title('Predicted vs. Actual House Prices', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('predicted_vs_actual.png', dpi=300)
plt.show()
```
